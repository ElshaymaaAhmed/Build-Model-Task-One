# Glucose Pathway 

```python
import cobra
```

```python
from cobra import Model, Reaction, Metabolite
```

```python
model = Model('task_one') # Create an object from class Model
```

### Create Reactions 

```python
v0 = Reaction('v0')
v0.name = 'v0'
v0.lower_bound = 1
v0.upper_bound = 1
```

```python
v1 = Reaction('v1')
v1.name = 'v1'
v1.lower_bound = 0
v1.upper_bound = 1000
```

```python
v2 = Reaction('v2')
v2.name = 'v2'
v2.lower_bound = 0
v2.upper_bound = 1000
```

```python
v3 = Reaction('v3')
v3.name = 'v3'
v3.lower_bound = 0.9
v3.upper_bound = 0.9
```

```python
v4 = Reaction('v4')
v4.name = 'v4'
v4.lower_bound = 0
v4.upper_bound = 1000
```

```python
M = Reaction('M')
M.name = 'M'
M.lower_bound = 0
M.upper_bound = 1000
```

### Create Metabolites 

```python
A = Metabolite('A', compartment = 'c')
B = Metabolite('B', compartment = 'c')
C = Metabolite('C', compartment = 'c')
ATP = Metabolite('ATP', compartment = 'c')
```

### Add Metabolites to Reactions 

```python
# V0 Reaction: enter to metabolite ---> A
v0.add_metabolites({A:1})
```

```python
# V1 Reaction: from metabolite A ---> B
v1.add_metabolites({A:-1,B:1})
```

```python
# V2 Reaction: from metabolite B ---> C
v2.add_metabolites({B:-1,C:1})
```

```python
# V3 Reaction: from metabolite A ---> ATP
v3.add_metabolites({A:-1,ATP:1})
```

```python
# V4 Reaction: from metabolite ATP ---> 
v4.add_metabolites({ATP:-1})
```

```python
# M Reaction: from metabolite C ---> 
M.add_metabolites({C:-1})
```

### Add Reactions to the Model 

```python
model.add_reactions([v0,v1,v2,v3,v4,M])
```

```python
model.objective = 'M'
```

```python
model.optimize()
```

```python
model.summary()
```

### Escher
